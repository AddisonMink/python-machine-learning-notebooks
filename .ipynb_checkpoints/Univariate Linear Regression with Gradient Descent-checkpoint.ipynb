# Univariate Linear Regression with Gradient Descent
In regression learning, we take a set of data points generated by some unknown function $f$ and try to infer a function $h$ that is similar to $f$.

In univariate linear regression in particular, try to approximate $f$ using a univariate linear function $h(x) = w_0 + w_1x$ where $w = [w_0, w_1]$ is a vector of real-numbered weights. So, our hypothesis space is isomorphic to $R^2$.

The first thing we need to determine is what constitutes a good hypothesis.

## The Loss Function
By convention, we'll use the mean-squared error loss function. Let $(x_1,y_1), (x_2,y_2), ..., (x_n,y_n)$ be our test set. Then our loss function is:

$$L(h_w) = {1 \over n}\sum_{i=1}^n (y_i - h_w(x_i))^2 = {1 \over n}\sum_{i=1}^n (y_i - (w_0 + w_1x))^2$$

In code:

```python
def meanSquaredError(X,Y,w0,w1):
    HX = w0 + X * w1
    ERR = (Y - HX) ** 2
    MSE = sum(ERR) / X.size
    return MSE

# The loss function is closure over the training set.
def mkLossFunc(X,Y):
    def L(w0,w1): 
        return meanSquaredError(X,Y,w0,w1)
    return L
```

Here's a plot of $L$ when $f(x) = 2 + 3x$.

```python
%matplotlib inline

from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

X = np.linspace(0, 10)
Y = 2 + X * 3
L = mkLossFunc(X,Y)

_w0 = np.linspace(0,5)
_w1 = np.linspace(0,5)
W0, W1 = np.meshgrid(_w0,_w1)

LOSS = []
for i in range(_w0.size):
    row = []
    for j in range(_w0.size):
        w0 = W0[i,j]
        w1 = W1[i,j]
        row.append(L(w0,w1))
    LOSS.append(row)
LOSS = np.array(LOSS)    
    
lossGraph = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(W0, W1, LOSS, 100, cmap='binary')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('L')
plt.show()    
```

As we can see in the contour plot, the value of $L$ is minimized $w_0 = 2$ and $w_1 = 3$.

Note that $L$ as convex. That means it has no local minima except for the global minimum. There is a unique weight vector, in this case $w^* = [2,3]$ where $L$ is minimized. 

The values of $w_0$ and $w_1$ that minimize $L$ will be the values at which ${\delta \over \delta w_0}L(h_w) = 0$ and ${\delta \over \delta w_1} L(h_w) = 0$.

Taking the partial derivatives, we get ${\delta \over \delta w_0}L(h_w) = {-2 \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x))$ and ${\delta \over \delta w_1}L(h_w) = {-2 \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x))x$.

The quations ${\delta \over \delta w_0}L(h_w) = {-2 \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x)) = 0$ and ${\delta \over \delta w_1}L(h_w) = {-2 \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x))x$ have unique solutions, so it's possible to infer $w^*$ directly, but we're going to use a more general approach to minimizing cost.

## Gradient Descent
$L$ is convex, so if we choose an arbitrary weight vector $w = [w_0, w_1]$ we can find $w^*$ with a greedy search.

First, we choose an arbitrary weight vector, say $w=[0,0]$. Then we compute $d_0={\delta \over \delta w_0} L(h_w)$ and $d_1 = {\delta \over \delta w_1} L(h_w)$

If $d_0 > 0$ then $w_0$ is too large, so we make it smaller. If $d_0 < 0$ then $w_0$ is too small, so we make it bigger. We do the same for $d_1$. Then we repeat this process with our new weight vector. If we repeat enough times we will eventually find $w^*$.

If we think about it graphically, a weight vector is a point on the contour plot of $L$. If the partial derivatives are positive, we are on the ascending slope, so we need to go backwards. If the partial derivatives are negative, we are on the descending slope, and we need to keep going forward. If the partial derivatives are 0, we are at the lowest point.

This algorithm is called gradient descent.

Before we can implement gradient descent, we need to make 2 of its componets more precise. How much should we update the weights at each step? How do we know when were are done?

### Learning Rate
At each step, we'll update our weight vector as follows:

$$w_0 := w_0 - \alpha {\delta \over \delta w_0} L(h_w) = w_0 + {2 \alpha \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x_i)$$

$$w_1 := w_1 - \alpha {\delta \over \delta w_1} L(h_w) = w_1 + {2 \alpha \over n} \sum_{i=1}^n (y_i - (w_0 + w_1x_i)x$$

Here $\alpha$ is a constant called the learning rate. It determines how much we update the weight vector at each step. If $\alpha$ is small, it will take a lot of steps to reach the bottom. If $\alpha$ is large, we might overshoot the bottom and start ascending the other side of the curve. If $\alpha$ is much too large, we'll keep overshooting the bottom and never converge. There is no _a priori_ way to know the best value for $\alpha$.

### Convergence Threshold
We know that $w^*$ exists, but we may not reach it in a finite number of steps so we'll stop the algorithm when we're close close to the $w^*$.

We'll know that we're close to $w^*$ when the gradient starts to flatten out. The closer we get to $w^*$, the smaller the reduction in cost between steps. At some point, the reduction in cost from an additional step will be so small that it's not worth taking. At that point we'll terminate the algorithm.

To put it precisely, we'll define some constant $\omega$ called the convergence threshold. If $w_i$ is the weight vector from the last step we took and $w_{i+1}$ is the weight vector from the current step, we will terminate the algorithm if:

$$|L(h_{w_i+1}) - L(h_{w_i})| < \omega$$

Like $\alpha$ there is no way to know the best value of $\omega$ beforehand. 

### Implementing Gradient Descent

```python
# The update functions are closures over the training set.
def mkW0UpdateFunc(X,Y,alpha):
    def updateW0(w0,w1):
        return w0 + (2 * alpha / X.size) * np.sum(Y - (w0 + w1 * X))
    return updateW0
        
def mkW1UpdateFunc(X,Y,alpha):
    def updateW1(w0,w1):
        return w1 + (2 * alpha / X.size) * np.sum((Y - (w0 + w1 * X)) * X)
    return updateW1
    
def gradientDescent(X,Y,alpha,omega):
    # Determine loss and update functions.
    L = mkLossFunc(X,Y)
    updateW0 = mkW0UpdateFunc(X,Y,alpha)
    updateW1 = mkW1UpdateFunc(X,Y,alpha)
    
    # Do algorithm.
    w0, w1 = 0, 0
    loss = L(w0,w1)
    while True:
        newW0 = updateW0(w0,w1)    
        newW1 = updateW1(w0,w1)
        w0, w1 = newW0, newW1    
        newLoss = L(w0,w1)
        if abs(newLoss - loss) < omega:
            break
        else:
            loss = newLoss
    return (w0,w1)
    

def fitLine(X,Y,alpha,omega):
    w0, w1 = gradientDescent(X,Y,alpha,omega)
    plt.scatter(X,Y)
    H = w0 + w1 * X
    plt.plot(X,H,'red')
    print("weights", w0, w1)
    print("loss:", mkLossFunc(X,Y)(w0,w1))
    return w0, w1
```

## Evaluating Gradient Descent on Known Functions
Here's the result of running gradient descent on 50 samples generated by the function $f(x) = 2 + 3x$ with $\alpha = 0.01$ and $\omega = 0.001$. I chose these values through trial and error.

```python
X = np.linspace(1,5)
Y = 2 + 3 * X
alpha = 0.01
omega = 0.001
fitLine(X,Y,alpha,omega)
```

Our hypothesis appears close to the scatterplot generated by $f$. Note that $w_0$ is too low and the difference is made up by making $w_1$ too high. Our hypothesis will be pretty accurate for values of $x$ within the bounds of the training set, it's accuracy will decrease the the more $x$ diverges from the bounds of the training set.

If our training set came from a real linear process, e.g. housing price based on square-footage, and our training set was large, then it would be rare for us to encounter new values of $x$ that were far outside the bounds of our test set. In that case, the inaccuracy of our hypothesis wouldn't matter very much.

Now let's look at an adversarial case where $f(x) = 2$. Technically, this fits our model because the weight vector $[2,0]$ is in our hypothesis space, but gradient descent doesn't do a very good job here.

```python
X = np.linspace(1,5)
Y = 2 + 0 * X
alpha = 0.01
omega = 0.001
fitLine(X,Y,alpha,omega)
```

The problem is that the update functions have know way of knowing which weight is responsible for the loss, so it tries to change both of them.

Here's another adversarial example where $f(x) = 2 + 3x + 3x^2$. This learning problem isn't realizable given our linear model, which shows that our choice of assumptions about $f$ has a big impact on our results.

```python
X = np.linspace(1,5)
Y = 2 + 3 * X + 3 * X * X
alpha = 0.01
omega = 0.001
fitLine(X,Y,alpha,omega)
```

## Evaluating Gradient Descent on Unknown Functions
Here's the result of running gradient descent on a set of datapoints comparing SAT score to high-school GPA. I randomly selected half the data points to use as a training set and reserved the other half to use as a test set.

If our hypothesis is good, it's loss on a test set it's never seen before should be similar to its loss on the training set.

```python
# randomly split scores in half
# use half as training, half as test.

import pandas as pd
import random

data = pd.read_csv("SAT-GPA.csv").values.T
print(data.size)
data = list(zip(data[0],data[1]))
random.shuffle(data)
data = np.array(data)
splitted = np.array_split(data,2)
train = splitted[0].T
test = splitted[1].T

trainX, trainY = train[0], train[1]
alpha = 0.000000001
omega = 0.000001
w0, w1 = fitLine(trainX,trainY,alpha,omega)
```

To measure the performance of our hypotheses, we'll see how well it predicts the test set.

```python
testX, testY = test[0], test[1]
HY = w0 + w1 * testX
plt.scatter(testX,testY)
plt.plot(testX,HY,'red')
print(mkLossFunc(testX,testY)(w0,w1))
```

Our hypothesis' loss on the test set is about 7% higher than its loss on the test set, so it generalizes fairly well.
